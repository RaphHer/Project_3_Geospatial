```python
#here I am making a query to get the gaming company located in france and I also want that their addresses are displayed

filter_ = {"offices.country_code": "FRA", "category_code": "games_video",}
projection = {"_id":0, "name":1,"offices":1,"category_code":1}
result = list(companies.find(filter_, projection))
df = pd.DataFrame(result)

df
```

```python
# Removing the [] so after I will be able to access the information inside it

df_1 = df.explode('offices')
df_1
```

```python
# Here I will create a new dataframe with the different element that were in the offices category 

Test=df_1['offices'].apply(pd.Series)
Test1=pd.concat([df, Test], axis=1)
Test1
```

```python
Test2 = Test1[Test1['latitude'].notna()]
```

```python
Test2.to_csv('1gaming_companies.csv', index = True, encoding='utf-8')
```

## Getting Advertising Companies in France

```python
filter_ = {"offices.country_code": "FRA", "category_code": "advertising",}
projection = {"_id":0, "name":1,"offices":1,"category_code":1}
result = list(companies.find(filter_, projection))
df_design = pd.DataFrame(result)
df_design
```

```python
df_design_2 = df_design.explode('offices')
df_design_2


Test_design = df_design_2['offices'].apply(pd.Series)
Test_design1 = pd.concat([df_design, Test_design], axis=1)
Test_design1
Test_design2 = Test_design1[Test_design1['latitude'].notna()]
Test_design2
Test_design3 = Test_design2[Test_design2['country_code'] == 'FRA']
Test_design3
```

```python
Test_design3.to_csv('2advertising_companies.csv', index = True, encoding='utf-8')
```

## Getting Companies that raised more than $1m in France

```python
filter_S = {"total_money_raised": {'$regex': "M|B"},"offices.country_code": "FRA"}
projection = {"_id":0, "name":1, "total_money_raised":1 ,"offices":1,"category_code":1}

result_S = list(companies.find(filter_S, projection))
df_startup = pd.DataFrame(result_S)
df_startup

df_startup_2 = df_startup.explode('offices')
df_startup_2


Test_startup = df_startup_2['offices'].apply(pd.Series)
Test_startup1 = pd.concat([df_startup, Test_startup], axis=1)
Test_startup1
Test_startup2 = Test_startup1[Test_startup1['latitude'].notna()]
Test_startup2
Test_startup3 = Test_startup2[Test_startup2['country_code'] == 'FRA']
Test_startup3

```

```python
Test_startup3.to_csv('3startup_companies.csv', index = True, encoding='utf-8')
```
