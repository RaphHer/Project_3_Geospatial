```python
#here I am making a query to get the gaming company located in france and I also want that their addresses are displayed

filter_ = {"offices.country_code": "FRA", "category_code": "games_video",}
projection = {"_id":0, "name":1,"offices":1,"category_code":1}
result = list(companies.find(filter_, projection))
df = pd.DataFrame(result)

df
```

```python
# Removing the [] so after I will be able to access the information inside it

df_1 = df.explode('offices')
df_1
```

```python
# Here I will create a new dataframe with the different element that were in the offices category 

Test=df_1['offices'].apply(pd.Series)
Test1=pd.concat([df, Test], axis=1)
Test1
```

```python
Test2 = Test1[Test1['latitude'].notna()]
```

## Putting these companies on a map 

```python
import folium
from folium import Choropleth, Circle, Marker, Icon, Map
from folium.plugins import HeatMap, MarkerCluster
import pandas as pd
```

```python
# coordinates of Franc 
france_location_lat = 46.22
france_location_long = 2.21
```

```python
#Show on the Map france
map_1 = Map(location=[france_location_lat, france_location_long], zoom_start = 5)
```

```python
map_1
```

```python
#Includes on the map the 
for i in range(0, len(Test2)):
    folium.Marker([Test2.iloc[i]["latitude"],Test2.iloc[i]["longitude"]],
                 popup = Test2.iloc[i]["name"], 
                  icon = Icon(
                    color='red',
                    opacity = 0.6,
                    prefix = "fa",
                    icon="gamepad",
                    icon_color = "black"    
                  )).add_to(map_1)
    
map_1
```
