# Visualisation  

```python
import csv
import folium
from folium import Choropleth, Circle, Marker, Icon, Map
from folium.plugins import HeatMap, MarkerCluster
import pandas as pd
```

## Gaming Companies

```python
df_Test_2 = pd.read_csv (r'1gaming_companies.csv')
```

```python
# coordinates of France 
france_location_lat = 46.22
france_location_long = 2.21
```

```python
#Show on the Map france
map_1 = Map(location=[france_location_lat, france_location_long], zoom_start = 5)
map_1
```

```python
#Includes on the map the 
for i in range(0, len(df_Test_2)):
    folium.Marker([df_Test_2.iloc[i]["latitude"],df_Test_2.iloc[i]["longitude"]],
                 popup = df_Test_2.iloc[i]["name"], 
                  icon = Icon(
                    color='red',
                    opacity = 0.6,
                    prefix = "fa",
                    icon="gamepad",
                    icon_color = "black"    
                  )).add_to(map_1)
    
map_1
```
