### 1. Kindergardens

```python
url_1 = "https://api.foursquare.com/v3/places/search?ll=48.86%2C2.34&categories=12056&sort=DISTANCE"

headers = {
    "accept": "application/json",
    "Authorization": "fsq33QFjrveyIY02656UccXwZF2gTUI9rGw3DxQymTbpy6c="
}

response = requests.get(url_1, headers=headers).json()
```

```python
response
```

```python
response.keys()
```

```python
response["results"][0]
```

```python
new_list = []
for i in response["results"]:
    
    name = i["name"]
    latitude = i["geocodes"]["main"]["latitude"]
    longitude = i["geocodes"]["main"]["longitude"]
        
    type_ = {"typepoint": 
                          {"type": "Point", 
                           "coordinates": [latitude, longitude]}}
    
    new_list.append({"name":name, "latitude":latitude, "longitude":longitude, "type":type_})
    
new_list
```

```python
df_kindergarden = pd.DataFrame(new_list)
df_kindergarden.sample()
```

```python
df_kindergarden.head()
```

```python
df_kindergarden.to_csv('4kindergarden.csv', index = True, encoding='utf-8')
```
